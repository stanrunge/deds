## Dataset

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler

df = pd.read_csv("student_habits_performance.csv")

df.drop(columns=["student_id"], inplace=True)

df = pd.get_dummies(df, drop_first=True)

X = df.drop("exam_score", axis=1)

y = df["exam_score"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

## Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

lr = LinearRegression()
lr.fit(X_train, y_train)
pred_lr = lr.predict(X_test)

mse_lr = mean_squared_error(y_test, pred_lr)
print("MSE Lineaire Regressie:", mse_lr)
```

## Neural Network

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense

nn = Sequential(
    [
        Dense(64, activation="relu", input_shape=(X_train.shape[1],)),
        Dense(32, activation="relu"),
        Dense(1),
    ]
)

nn.compile(optimizer="adam", loss="mse")
nn.fit(X_train, y_train, epochs=50, batch_size=16, validation_split=0.1)

pred_nn = nn.predict(X_test).flatten()
mse_nn = mean_squared_error(y_test, pred_nn)
print("MSE Neuraal Netwerk:", mse_nn)
```

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping

nn = Sequential(
    [
        Dense(128, activation="relu", input_shape=(X_train.shape[1],)),
        BatchNormalization(),
        Dropout(0.3),
        Dense(64, activation="relu"),
        BatchNormalization(),
        Dropout(0.3),
        Dense(32, activation="relu"),
        Dense(1),
    ]
)

nn.compile(optimizer=Adam(learning_rate=0.001), loss="mse")

early_stop = EarlyStopping(monitor="val_loss", patience=10, restore_best_weights=True)

history = nn.fit(
    X_train,
    y_train,
    epochs=200,
    batch_size=16,
    validation_split=0.1,
    callbacks=[early_stop],
)

pred_nn = nn.predict(X_test).flatten()
mse_nn = mean_squared_error(y_test, pred_nn)
print("Nieuwe NN - MSE:", mse_nn)
```
